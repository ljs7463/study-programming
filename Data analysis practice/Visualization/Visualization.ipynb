# 전체 스케일 조정
sns.set_context()안에 4종류의 스케일(사이즈)을 선택할 수 있다. 
"paper", "notebook", "talk", "poster"
여기서 기본값은 notebook 이다. 

가장 작은 스케일 = paper
가장 큰 스케일 = poster


```python
plt.figure(figsize = (12,6))
sns.set_style('whitegrid')
sns.set_context('talk')

sns.countplot(data['Contract'], hue = data['Churn'])
```

# 폰트 사이즈 조정
스케일로도 전반적인 사이즈를 조정할 수 있지만 실제 폰트 크기를 조정하는 파라미터는 별도로 있다.
sns.set_context()안에 font_scale을 넣어주면 된다. 

```python
plt.figure(figsize = (12,6))
sns.set_style('whitegrid')
sns.set_context('talk',font_scale = 1)

sns.countplot(data['Contract'], hue = data['Churn'])
```

# rc파라미터로 세부 조정
만약 디테일하게 조정하고 싶다면 sns.set_context()안에 rc파라미터를 딕셔너리 형태로 넣어 주면 된다.
rc = Run Command의 약자이다


```python
plt.figure(figsize = (12,6))
sns.set_style('whitegrid')
sns.set_context('talk',font_scale = 1,rc = {"grid.linewidth" : 5})

sns.countplot(data['Contract'], hue = data['Churn'])
```
