```
# data 생성
data = np.arange(1, 51)

# 밑 그림
fig, axes = plt.subplots(2, 3)

axes[0, 0].plot(data)
axes[0, 1].plot(data * data)
axes[0, 2].plot(data ** 3)
axes[1, 0].plot(data % 10)
axes[1, 1].plot(-data)
axes[1, 2].plot(data // 20)


plt.tight_layout()
plt.show()
```

## 1-1. x, y, colors, area 설정하기
  - colors 는 임의 값을 color 값으로 변환합니다.
  - area는 점의 넓이를 나타냅니다. 값이 커지면 당연히 넓이도 커집니다.

```
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.arange(50)  
area = x * y * 1000  #area = 면적 즉 값의 크기에 따라 달라진다
```

```
print(colors)
```

```
plt.scatter(x, y, s = area, c = colors) # s = size
plt.show()

# float 은 0 ~ 1 까지, 정수는 0 ~ 255 까지를 색으로 표현할 수 있다.
```

## 1-2. cmap과 alpha
  - cmap에 컬러를 지정하면, 컬러 값을 모두 같게 가져갈 수도 있습니다.
  - alpha값은 투명도를 나타내며 0 ~ 1 사이의 값을 지정해 줄 수 있으며, 0에 가까울 수록 투명한 값을 가집니다.
  

```
np.random.rand(50)
```

```
plt.figure(figsize = (12, 6))

plt.subplot(131)
plt.scatter(x, y, s = area, cmap = 'blue', alpha = 0.1)
plt.title('alpha = 0.1')

plt.subplot(132)
plt.title('alpha = 0.5')
plt.scatter(x, y, s = area, cmap = 'blue', alpha = 0.5)

plt.subplot(133)
plt.title('alpha = 1.0')
plt.scatter(x, y, s = area, cmap = 'blue', alpha = 1.0)

plt.show()
```

# 2. Barplot, Barhplot

1개의 canvas 안에 다중 그래프 그리기

## 2-1. 기본 Barplot 그리기

```
x = ['Math', 'Programming', 'DataScience', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

#plt.figure(figsize = (6, 3)) 이 코드 작성시 xtick들의 내용이 서로겹침 그땐 rotation을 써도 되고 아니면 Barh그래프로 변경
plt.bar(x, y, align = 'center', alpha = 0.6, color = 'red') # align = 'center' 은 tick을 center로 표시해라 <->(align = 'edge')
plt.xticks(x)
plt.ylabel('Number of Students',fontsize = 20)
plt.title('Subjects')

plt.show()
```

## 2-2. 기본 Barhplot 그리기

barh 함수에서는 xticks로 설정했던 부분을 yticks로 변경 합니다.

```
x = ['Math', 'Programming', 'DataScience', 'Art', 'English', 'Physics']
y = [66, 80, 60, 50, 80, 10]

plt.barh(x, y, align = 'center', alpha = 0.7, color = 'green')
plt.yticks(x)
plt.xlabel('Number of Students')
plt.title('Subgects')

plt.show()
```

## Barplot에서 비교 그래프 그리기

동일한 과목 두학생 비교하고 싶을때

```
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label)) #x는 xticks에 넣을거니까 당연이 x_label과 같은 갯수여야 한다 따라서 대부분 이 코드 사용
print(x)
y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

# 넓이 지정
width = 0.35

# subplots 생성
fig, axes = plt.subplots()


# 넓이 설정
axes.bar(x - width/2, y_1, width, align = 'center', alpha = 0.5)
 # x-width/2 만큼 왼쪽으로 가서 width만큼의 넓이의 bar그래프를 그리는데  
 # align = 'center' 니까 x좌표(x-width/2)가 중앙으로 해서 width만큼의 bar그래프를 그려준다
axes.bar(x + width/2, y_2, width, align = 'center', alpha = 0.8)



# xtick 설정
plt.xticks(x)
axes.set_xticklabels(x_label)
plt.ylabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])
plt.show()
```

```
x_label = ['Math', 'Programming', 'Data Science', 'Art', 'English', 'Physics']
x = np.arange(len(x_label))

y_1 = [66, 80, 60, 50, 80, 10]
y_2 = [55, 90, 40, 60, 70, 20]

# 넓이 지정
width = 0.35

# subplots 생성
fig, axes = plt.subplots()

# 넓이 설정
axes.barh(x - width/2, y_1, width, align = 'center', alpha = 0.5, color = 'green')
axes.barh(x + width/2, y_2, width, align = 'center', alpha = 0.8, color = 'red')

# xtick 설정
plt.yticks(x)
axes.set_yticklabels(x_label)
plt.xlabel('Number of Students')
plt.title('Subjects')

plt.legend(['john', 'peter'])
plt.show()
```

# 3. Line Plot

## 3-1. 기본 lineplot그리기

```
x = np.arange(0, 10, 0.1)
y = 1 + np.sin(x)
#print(x)
plt.plot(x, y)

plt.xlabel('x value', fontsize = 15)
plt.ylabel('y value', fontsize = 15)
plt.title('sin graph', fontsize = 18)

plt.grid()
plt.show()
```

## 3-2. 2개 이상의 그래프 그리기


#####  - color : 컬러 옵션
#####  - alpha : 투명도 옵션

```
x = np.arange(0, 10, 0.1)
y_1 = 1 + np.sin(x)
y_2 = 1 + np.cos(x)

plt.plot(x, y_1, label = '1 + sin', color = 'blue', alpha = 0.3)
plt.plot(x, y_2, label = '1 + cos', color = 'red', alpha = 0.7)

plt.xlabel('x value', fontsize = 15)
plt.ylabel('y value', fontsize = 15)
plt.title('sin and cos graph', fontsize = 18)

plt.grid()
plt.legend(fontsize = 12)

plt.show()
```

## 3-3. 마커 스타일링

# 5. Histogram

## 5-1. 기본 Histogram 그리기

```
N = 100000 # 데이터 10만개
bins = 30 # 구간

x = np.random.randn(N)

plt.hist(x, bins = bins)

plt.show()

# 데이터의 분포를 보고자 할 때 히스토그램을 많이 사용한다.
```

  - sharey : y축을 다중 그래프가 share ( subplots를 그릴건데 y축을 일일이 다표기하기 번거로워서 y축을share해서 한개로 퉁치겠다)
  - tight_layout : graph의 패딩을 자동으로 조절해주어 fit한 graph를 생성

## 5-2. 다중 Histogram 그리기

```
N = 100000 
bins = 30

x = np.random.randn(N)

fig, axs = plt.subplots(1, 3,
                        sharey = True,
                        tight_layout = True
                        )

fig.set_size_inches(12, 5)

axs[0].hist(x, bins = bins)
axs[1].hist(x, bins = bins * 2)
axs[2].hist(x, bins * 4)

plt.show()
```

## 5-3. Y축에 Density 표기

```
N = 100000 
bins = 30

x = np.random.randn(N)

fig, axs = plt.subplots(1, 2,
                        tight_layout = True)

fig.set_size_inches(9, 3)

# density = True 값을 통하여 Y축에 density(=밀도)를 표기할 수 있습니다.(실질적인 개수가 아닌 몇 퍼샌트가 몰려있는지 )
axs[0].hist(x, bins = bins, density = True, cumulative = True) # cumulative 는 누적분포로 본다.
axs[1].hist(x, bins = bins, density = True)


plt.show()
```

# 6. Pie Chart

#####점유율을 표기할 때 많이 사용한다. 다만, 점유율 표기할때 4, 5개 는 적절하지만 20, 30개는 부적절하다 
이럴땐 차라리 bar차트 사용

### pie chart 옵션
  
  - explode : 파이에서 툭 튀어져 나온 비율
  - autopct : 퍼센트 자동으로 표기
  - shadow : 그림자 표시
  - startangle : 파이를 그리기 시작할 각도

texts , autotexts 인자를 리턴 받습니다.

texts는 label에 대한 텍스트 효과를

autotexts는 파이 위에 그려지는 텍스트 효과를 다룰 때 활용합니다.

```
labels = ['Samsung', 'Huawei', 'Apple', 'Xiaomi', 'Oppo', 'Etc']
sizes = [20.4, 15.8, 10.5, 9, 7.6, 36.7 ]
explode = (0.3, 0, 0, 0, 0, 0)

# texts, autotexts 인자를 활용하여 텍스트 스타일링을 적용합니다.
patches, texts, autotexts = plt.pie(sizes,
                                    explode = explode,
                                    labels = labels,
                                    autopct = '%1.1f%%', # .1f 는 소수점 몇째 자리
                                    shadow = True,
                                    startangle = 90)
# patches는 신경쓰지말고 texts 와 autotexts를 인자로 받는데 이것으로 추가 옵션을 줄 수 있다.

plt.title('Smartphone pie', fontsize = 15)

# label 텍스트에 대한 스타일 적용
for t in texts:
  t.set_fontsize(12)
  t.set_color('gray')

# pie 위의 텍스트에 대한 스타일 적용
for t in autotexts:
  t.set_color("white")
  t.set_fontsize(18)

plt.show()
```

```
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np


fig = plt.figure(figsize = (10, 5))
ax = fig.add_subplot(111, projection = '3d') # Axe3D object

sample_size = 500

x = np.cumsum(np.random.normal(0, 5, sample_size))

y = np.cumsum(np.random.normal(0, 5, sample_size))

z = np.cumsum(np.random.normal(0, 5, sample_size))

ax.scatter(x, y, z, c = z, s = 20, alpha = 0.5, cmap = 'Greens')

plt.title('ax.scatter')
plt.show()
```

# 9. imshow

이미지(image)데이터와 유사하게 행과 열을 가진 2차원의 데이터를 시각화 할 떄는 imshow를 활용한다.

```
from sklearn.datasets import load_digits

digits = load_digits()
X = digits.images[:10]
X[0]
```

  - load_digits는 0 ~ 16 값을 가지는 array로 이루어져 있습니다. # 0은 흰, 16은 진하다
  - 1개의 array는 8 x 8 배열 안에 표현되어 있습니다.
  - 숫자는 0 ~ 9 까지 이루어져 있습니다.

```
fig, axes = plt.subplots(nrows = 2, ncols = 5, figsize = (12, 6), sharex = True,  sharey = True)

for i in range(10):
  axes[i//5][i%5].imshow(X[i], cmap = 'Blues')
  axes[i//5][i%5].set_title( str(i), fontsize = 20)

plt.tight_layout()
plt.show()
```
